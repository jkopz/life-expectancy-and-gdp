# Importing Python Modules

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading the data and inspecting the DataFrame

```python
df = pd.read_csv("all_data.csv")
print(df.head())
```

Changing the column name `"Life expectancy at birth (years)"` to `"life_expactancy"`

```python
df.rename(columns={"Life expectancy at birth (years)": "life_expactancy"}, inplace = True)
print(df.head())
print(df.Country.unique())
```

# Examining the data

The dataset provided has the following columns of data:
* **Country** = nation
* **Year** = the year for the observation
* **life_expactancy** = life expectancy value in years
* **GDP** = Gross Domestic Product in U.S. dollars

The GDP and life expactancy of the following countries will be examined:
* Chile
* China
* Germany
* Mexico
* USA
* Zimbabwe

The datasets contains data from 2000 to 2015.

# Exploring the evolution of life expactancy and the GDP for every country

Instead of plotting every country on its own, I made a function which will then create all plots at once.

```python
def gdp_life_expactancy(country):
    df_country = df[df.Country == country]
    ax = plt.subplot()
    df_country.plot(x="Year", y="GDP", ax=ax)
    df_country.plot(x="Year", y="life_expactancy", ax=ax, secondary_y=True)
    ax.set_xticks(df_country.Year.unique())
    ax.set_xticklabels(df_country.Year.unique(), rotation=30)
    ax.set_ylabel("GDP in Trillion $")
    ax.right_ax.set_ylabel("Life expectancy in years")
    ax.set_title(country.title())
    return plt.show()

# Extracting the countries to use them as parameters
countries = df["Country"].unique()

# Creating six figures for each plot
fig, axes = plt.subplots(6,1) 

# Zipping the countries and the axes
for country, ax in zip(countries, axes):
    ax = gdp_life_expactancy(country)
    

plt.tight_layout()
plt.show()
```
